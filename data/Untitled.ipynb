# Pre-process CO2 emission data

```python
#read the csv
import pandas as pd
df_CO2 = pd.read_csv('C:/Users/ryuya/Desktop/FIT3179_Visual Studio/FIT3179_Assignment2/data/CO2 emission by countries.csv',
                  encoding='latin1')
df_CO2.head()
```

```python
df_CO2['Country'].unique()
```

```python
#Change the country name
df_CO2.loc[df_CO2['Country'] == 'United States', 'Country'] = 'United States of America'

df_CO2.loc[df_CO2['Country'] == 'Democratic Republic of Congo', 'Country'] = 'Dem. Rep. Congo'

df_CO2.loc[df_CO2['Country'] == 'Central African Republic', 'Country'] = 'Central African Rep.'

df_CO2.loc[df_CO2['Country'] == 'South Sudan', 'Country'] = 'S. Sudan'

df_CO2.loc[df_CO2['Country'] == 'Equatorial Guinea', 'Country'] = 'Eq. Guinea'
```

```python
#rename the column name
df_CO2.rename(columns = {'CO2 emission (Tons)':'CO2_emission_tons', 
                        'Population(2022)':'Population_2022',
                        '% of World':'%_of_world',
                        'Density(km2)':'density_km2'}, 
                        inplace = True)
df_CO2.head()
```

```python
#Filter data with year 2020
df_CO2_2020 = df_CO2.loc[df_CO2['Year'] == 2020]
```
